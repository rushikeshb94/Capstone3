```python
%%writefile app.py
#importing all the libraries
import io
import os
import scipy.misc
import numpy as np
import six
import time
import glob
from IPython.display import display

from six import BytesIO

import matplotlib
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont

import tensorflow as tf
from object_detection.utils import ops as utils_ops
from object_detection.utils import label_map_util
from object_detection.utils import visualization_utils as vis_util

import streamlit as st

from inference_utils import *

@st.cache(allow_output_mutation=True)
#method to load the model
def load_model():
  tf.keras.backend.clear_session()
  model = tf.saved_model.load(f'/content/inference_graph/saved_model/')
  return model

with st.spinner('Model is being loaded..'):
  model=load_model()

st.write("""
         # Aquarium Images Object Detection with Tensor Object Detection API 
         """
         )

file = st.file_uploader("Upload the image to be classified U0001F447", type=["jpg", "png"])
st.set_option('deprecation.showfileUploaderEncoding', False)

category_index = label_map_util.create_category_index_from_labelmap('/content/labelmap.pbtxt', use_display_name=True)

def upload_predict(upload_image, model):
      #upload_image = Image object
      #model = loaded model
      (im_width, im_height) = image.size
      image_np = np.array(image.getdata()).reshape((im_height, im_width, 3)).astype(np.uint8)
      output_dict = run_inference_for_single_image(model, image_np)
      vis_util.visualize_boxes_and_labels_on_image_array(
      image_np,
      output_dict['detection_boxes'],
      output_dict['detection_classes'],
      output_dict['detection_scores'],
      category_index,
      instance_masks=output_dict.get('detection_masks_reframed', None),
      use_normalized_coordinates=True,
      line_thickness=8)
      pred = Image.fromarray(image_np)
        
      return pred

if file is None:
    st.text("Please upload an image file")
else:
    image = Image.open(file)
    st.image(image, use_column_width=True)
    predictions = upload_predict(image, model)
    st.write('The image detection results below')
    st.image(predictions)
```
